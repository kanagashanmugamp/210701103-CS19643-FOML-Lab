```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt # plotting
import numpy as np # linear algebra
import os # accessing directory structure
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
data=pd.read_csv("/content/train.csv")
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x_data=data.iloc[:10,0:1].values
y_data=data.iloc[:10,1:2].values
```

```python
model = LinearRegression()
model.fit(x_data,y_data)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x_data,y_data,color="red")  #scatter plot to know where the data points are actually located
plt.plot(y_data,model.predict(y_data),color='green') #plotting the linear line
plt.title("Linear regression")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)

diabetes_X = diabetes_X[:, np.newaxis, 2]

diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]

diabetes_y_train = diabetes_y[:-20]
diabetes_y_test = diabetes_y[-20:]

regr = linear_model.LinearRegression()

regr.fit(diabetes_X_train, diabetes_y_train)


diabetes_y_pred = regr.predict(diabetes_X_test)
plt.scatter(diabetes_X_test, diabetes_y_test, color="red")
plt.plot(diabetes_X_test, diabetes_y_pred, color="blue", linewidth=3)
plt.title("Diabetes Dataset-Linear regression")
plt.xlabel('x')
plt.ylabel('y')

# plt.xticks(())
# plt.yticks(())

plt.show()
```
